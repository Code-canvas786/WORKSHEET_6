```python
import matplotlib.pyplot as plt

students = ['Ann', 'Aditya', 'Chirag', 'Gurleen', 'Kunal']
math = [85, 79, 90, 66, 70]
physics = [78, 82, 85, 75, 68]
chemistry = [92, 74, 80, 88, 75]
english = [88, 90, 92, 78, 85]

total_marks = [sum(scores) for scores in zip(math, physics, chemistry, english)]
average_marks = [total / 4 for total in total_marks]

top_performer_index = total_marks.index(max(total_marks))
bottom_performer_index = total_marks.index(min(total_marks))

print(f"Top Performer: {students[top_performer_index]} with {total_marks[top_performer_index]} marks")
print(f"Bottom Performer: {students[bottom_performer_index]} with {total_marks[bottom_performer_index]} marks")

x = range(len(students))
plt.bar(x, math, width=0.2, label='Math', align='center')
plt.bar([i + 0.2 for i in x], physics, width=0.2, label='Physics', align='center')
plt.bar([i + 0.4 for i in x], chemistry, width=0.2, label='Chemistry', align='center')
plt.bar([i + 0.6 for i in x], english, width=0.2, label='English', align='center')
plt.xticks([i + 0.3 for i in x], students)
plt.title('STUDENT PERFORMANCE')
plt.xlabel('STUDENTS')
plt.ylabel('MARKS')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

time = np.array([0, 1, 2, 3, 4, 5])
velocity = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

def quadratic(t, a, b, c):
    return a*t**2 + b*t + c

params, _ = curve_fit(quadratic, time, velocity)
a, b, c = params

t_fit = np.linspace(0, 5, 100)
v_fit = quadratic(t_fit, a, b, c)

plt.scatter(time, velocity, color='green', label='Original Data')
plt.plot(t_fit, v_fit, label=f'Fitted Curve: v(t) = {a:.2f}t + {b:.2f}t + {c:.2f}')
plt.xlabel('Time (seconds)')
plt.ylabel('Velocity (m/s)')
plt.legend()
plt.title('Velocity Curve Fitting')
plt.show()
```

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

years = [2000, 2005, 2010, 2015]
population = [50, 55, 60, 70]

linear_interp = interp1d(years, population, kind='linear')

year_to_estimate = 2008
estimated_population = linear_interp(year_to_estimate)
print(f"Estimated population for the year 2008: {estimated_population:.2f} thousand")

years_fit = np.linspace(2000, 2015, 100)
population_fit = linear_interp(years_fit)

plt.scatter(years, population, color='green', label='Original Data')
plt.plot(years_fit, population_fit, label='Linear Interpolation')
plt.scatter(year_to_estimate, estimated_population, color='violet', label=f'Estimate for {year_to_estimate}')
plt.xlabel('Year')
plt.ylabel('Population (thousands)')
plt.legend()
plt.title('Population Interpolation')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

coeffs = [3, -5, 2, -8]
roots = np.roots(coeffs)
print(f"Roots of the polynomial: {roots}")

x = np.linspace(-3, 3, 400)
y = np.polyval(coeffs, x)

plt.plot(x, y, label='p(x) = 3x - 5x + 2x - 8')
plt.axhline(0, color='black', lw=0.5)
plt.scatter(roots, [0]*len(roots), color='red', zorder=5, label='Roots')
plt.xlabel('x')
plt.ylabel('p(x)')
plt.legend()
plt.title('Polynomial Plot and Roots')
plt.show()
```

```python
import time

file_sizes = [200, 400, 600, 800, 1000]
results = {}

for size in file_sizes:
    filename = f'test_file_{size}MB.txt'

    with open(filename, 'w') as f:
        f.write("a" * (size * 1024 * 1024))

    start_time = time.time()

    with open(filename, 'r') as f:
        content = f.read()
        content_upper = content.upper()


    with open(f'uppercase_{filename}', 'w') as f:
        f.write(content_upper)


    elapsed_time = time.time() - start_time
    results[size] = elapsed_time
    print(f"Time taken for {size} MB file: {elapsed_time:.2f} seconds")


import matplotlib.pyplot as plt

plt.plot(list(results.keys()), list(results.values()), marker='o')
plt.xlabel('File Size (MB)')
plt.ylabel('Time (seconds)')
plt.title('File Conversion Performance')
plt.show()
```

```python
from scipy.optimize import minimize_scalar
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**4 - 3*x**2 + 2

result = minimize_scalar(f, bounds=(-2, 3), method='bounded')
print(f"Local minimum at x = {result.x}, f(x) = {result.fun}")

x = np.linspace(-2, 3, 400)
y = f(x)

plt.plot(x, y, label='f(x) = x - 3x + 2')
plt.scatter(result.x, result.fun, color='violet', label=f'Local Minima at x = {result.x:.2f}')
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Function Plot and Local Minima')
plt.show()
```
